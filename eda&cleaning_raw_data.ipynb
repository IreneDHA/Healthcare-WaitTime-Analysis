Moving on to cleaning fase

```python
# Create a new temporary column 'Visit Provisional' by converting 'Visit Date' to datetime format
# This allows proper extraction of date and time components using pandas datetime methods
wait['Visit Provisional'] = pd.to_datetime(wait['Visit Date'])
```

```python
# Replace the original 'Visit Date' column with only the date portion (time set to 00:00:00)
# .normalize() ensures we get a clean date without time component
wait["Visit Date"] = wait["Visit Provisional"].dt.normalize() 
```

```python
# Create a new 'Visit Time' column containing only the time portion
wait["Visit Time"] = wait["Visit Provisional"].dt.strftime("%H:%M:%S") 
```

```python
# Remove the temporary 'Visit Provisional' column as it's no longer needed
wait.drop(columns= 'Visit Provisional', inplace=True)
```

```python
# Display the data types of all columns to verify the transformations were successful
wait.dtypes
```

```python
#Quick visualization of the new columns and data
wait.head()
```

```python
# Get all column names from the DataFrame and convert them to a list
cols = wait.columns.tolist()

# Move "Visit Time" column to appear right after "Visit Date" column:
cols.insert(cols.index("Visit Date") + 1, cols.pop(cols.index("Visit Time")))

# Reorder the DataFrame columns using the modified column list
wait = wait[cols]
```

```python
# This shows the current order and names of all columns after the reorganization
wait.columns
```

```python
# This verifies data consistency: 'Total Wait Time' should equal the sum of its parts
time_mask = (
    wait['Total Wait Time (min)'] ==
    wait['Time to Registration (min)'] +
    wait['Time to Triage (min)'] +
    wait['Time to Medical Professional (min)'])
```

```python
# Check if ALL rows in the mask are True (meaning all records are consistent)
time_mask.all()
```

```python
# Count the number of inconsistent records:
(~time_mask).sum()
```

```python
# Calculate number of patients per date and time of day combination
# transform('count') ensures the count is broadcast back to each row in the original DataFrame
wait['Number of Patients'] = wait.groupby(['Visit Date', 'Time of Day'])['Visit ID'].transform('count')
```

```python
# Calculate the specialist-to-patient ratio
wait['Specialist-to-Patient Ratio'] = (wait['Specialist Availability'] / wait['Number of Patients']).round(2)
```

```python
# Display all unique values in the 'Specialist-to-Patient Ratio' column to check for any unexpected results
wait['Specialist-to-Patient Ratio'].unique()
```

```python
# Get current column names as a list for reordering
cols = wait.columns.tolist()

# Move "Specialist-to-Patient Ratio" column right after "Specialist Availability":
cols.insert(cols.index('Specialist Availability') + 1, cols.pop(cols.index('Specialist-to-Patient Ratio')))

# Reorder the DataFrame columns using the modified column list
wait = wait[cols]
```

```python
# Display the final column order to verify the reorganization
wait.columns
```

```python
# Export the cleaned DataFrame to a CSV file in the 'files' directory
wait.to_csv('files/ER_wait_time_clean.csv')
```
